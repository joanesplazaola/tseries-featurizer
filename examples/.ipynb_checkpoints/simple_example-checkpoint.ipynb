```python
from ts_featurizer import TimeSeriesFeaturizer
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime

%matplotlib inline
```

```python
N = 3000

signal = np.sin(50 * np.pi * np.arange(N) / float(N / 2)) * 21
signal2 = np.sin(5 * np.pi * np.arange(N) / float(N / 2)) * 20
noise = np.random.normal(0, 1, N )
signal = signal + signal2  + noise
base = datetime.datetime(2000, 1, 1)
time = np.array([base + datetime.timedelta(hours=i) for i in range(N)])
```

```python

d = {
	'sensor1': signal,
	'sensor2': signal2 + noise,
	'sensor3': noise,
	'time': time

}

data = pd.DataFrame(d)
data
```

```python
data[['sensor1', 'sensor2', 'sensor3']].plot()
```

```python

tseries = TimeSeriesFeaturizer()

transformed = tseries.featurize(data, n_jobs=16)
transformed
```

```python
test_features = tseries.test_featurization(data, n_jobs=16)
test_features
```
